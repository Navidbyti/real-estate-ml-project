```python
# ==============================
# Helper: find columns to drop
# ==============================

def get_drop_columns(df, min_fill_ratio=1.0, always_drop=None):
    if always_drop is None:
        always_drop = []

    n = len(df)

    # Columns not fully filled
    not_full = [
        col for col in df.columns
        if df[col].notna().sum() < n
    ]

    # Very sparse columns
    very_sparse = [
        col for col in df.columns
        if df[col].notna().sum() / n < min_fill_ratio
    ]

    drop_cols = set(not_full + very_sparse + always_drop)
    return sorted(drop_cols)


# ==============================
# BUY DATAFRAME CLEANING
# ==============================

always_drop_buy = [
    "Unnamed: 0",
    "description",
    "title",
    "location_radius",
    "property_type"
]

drop_buy = get_drop_columns(
    df_buy,
    min_fill_ratio=1.0,
    always_drop=always_drop_buy
)

df_buy_clean = df_buy.drop(columns=drop_buy)


# ==============================
# RENT DATAFRAME CLEANING
# ==============================

always_drop_rent = [
    "Unnamed: 0",
    "description",
    "title",
    "rent_to_single",
    "rent_type",
    "deed_type",
    "transformed_credit",
    "transformed_rent",
    "property_type",
    "location_radius"
]

drop_rent = get_drop_columns(
    df_rent,
    min_fill_ratio=1.0,
    always_drop=always_drop_rent
)

df_rent_clean = df_rent.drop(columns=drop_rent)


# ==============================
# DAILY RENT DATAFRAME CLEANING
# ==============================

always_drop_daily = [
    "Unnamed: 0",
    "description",
    "title",
    "land_size",
    "deed_type",
    "floor",
    "total_floors_count",
    "unit_per_floor",
    "construction_year",
    "is_rebuilt",
    "building_direction",
    "floor_material",
    "location_radius"
]

drop_daily = get_drop_columns(
    df_daily,
    min_fill_ratio=1.0,
    always_drop=always_drop_daily
)

df_daily_clean = df_daily.drop(columns=drop_daily)


# ==============================
# FINAL SANITY CHECK
# ==============================

for name, df_ in {
    "BUY": df_buy_clean,
    "RENT": df_rent_clean,
    "DAILY": df_daily_clean
}.items():
    print(f"\n{name} max nulls in any column:")
    print(df_.isna().sum().max())
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder



# -------------------------------
# TARGETS
# -------------------------------
y_buy = df_buy_clean['price_value'].copy()
y_rent = df_rent_clean['target_credit'].copy()  # converted credit target
y_daily = df_daily_clean['rent_price_on_regular_days'].copy()

# -------------------------------
# DROP TARGET-LEAKING COLUMNS
# -------------------------------
df_buy_final = df_buy_clean.drop(columns=['price_value', 'price_mode'])
df_rent_final = df_rent_clean.drop(columns=[
    'rent_value', 'credit_value', 'rent_mode', 'credit_mode', 
    'rent_credit_transform', 'transformable_price', 'transformable_credit'
])
df_daily_final = df_daily_clean.drop(columns=['rent_price_on_regular_days'])

# -------------------------------
# IDENTIFY NUMERIC AND CATEGORICAL FEATURES
# -------------------------------
def get_features(df, exclude=[]):
    numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    # Remove excluded columns
    numeric_cols = [c for c in numeric_cols if c not in exclude]
    categorical_cols = [c for c in categorical_cols if c not in exclude]
    return numeric_cols, categorical_cols

# Example: don't scale/drop ID columns
numeric_buy, categorical_buy = get_features(df_buy_final, exclude=['Unnamed: 0'])
numeric_rent, categorical_rent = get_features(df_rent_final, exclude=['Unnamed: 0'])
numeric_daily, categorical_daily = get_features(df_daily_final, exclude=['Unnamed: 0'])

# -------------------------------
# SCALER AND ENCODER
# -------------------------------
def preprocess_df(df, numeric_cols, categorical_cols):
    df_processed = df.copy()
    
    # Scale numeric
    if numeric_cols:
        scaler = StandardScaler()
        df_processed[numeric_cols] = scaler.fit_transform(df_processed[numeric_cols])
    
    # Encode categoricals
    for col in categorical_cols:
        if df_processed[col].nunique() <= 10:  # small cardinality → LabelEncoder
            le = LabelEncoder()
            df_processed[col] = le.fit_transform(df_processed[col].astype(str))
        else:  # large cardinality → OneHotEncoder
            ohe = pd.get_dummies(df_processed[col].astype(str), prefix=col)
            df_processed = pd.concat([df_processed.drop(columns=[col]), ohe], axis=1)
    
    return df_processed

# -------------------------------
# PROCESS FINAL DATAFRAMES
# -------------------------------
df_buy_final = preprocess_df(df_buy_final, numeric_buy, categorical_buy)
df_rent_final = preprocess_df(df_rent_final, numeric_rent, categorical_rent)
df_daily_final = preprocess_df(df_daily_final, numeric_daily, categorical_daily)

# -------------------------------
# READY TO USE
# -------------------------------
print(df_buy_final.shape, y_buy.shape)
print(df_rent_final.shape, y_rent.shape)
print(df_daily_final.shape, y_daily.shape)
```
